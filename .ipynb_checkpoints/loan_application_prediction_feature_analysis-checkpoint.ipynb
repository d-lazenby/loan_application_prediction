```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

sns.set_palette("Set2")

df = pd.read_csv('Loan-Approval-Prediction.csv')
```

```python
df['TotalIncome'] = df['ApplicantIncome'] + df['CoapplicantIncome']
```

```python
df['TotalIncome'].hist(bins=40)

plt.ylabel('Count')
plt.xlabel('Total income');
```

```python
categorical = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Credit_History', 
               'Property_Area']

numerical = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'TotalIncome', 'Loan_Amount_Term']
```

```python
df.isna().sum()
```

### Numerical features

```python
from sklearn.feature_selection import f_classif
```

```python
mask = (df[numerical[0]].notna() & 
        df[numerical[1]].notna() & 
        df[numerical[2]].notna() & 
        df[numerical[3]].notna())

df_anova = df[mask][numerical[:4] + ['Loan_Status']]
y_anova = df_anova.pop('Loan_Status')
```

```python
[f_stats, f_p_values] = f_classif(df_anova, y_anova)
```

```python
f_test = pd.DataFrame({'features': numerical[:4],
                      'f-score': f_stats,
                      'p-values': f_p_values})
```

```python
f_test
```
