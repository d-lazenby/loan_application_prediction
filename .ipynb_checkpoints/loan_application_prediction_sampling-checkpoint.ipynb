```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("Set2")

df = pd.read_csv('Loan-Approval-Prediction.csv')

categorical = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Credit_History', 
               'Property_Area']

numerical = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
```

```python
df.info()
```

### Split into training and test sets

Ideally, to avoid data leakage, no data from the test set should be used in preprocessing, so we'll split into training and test sets beforehand. Given that we have less than 1000 rows, we'll have a relatively small test set and make an 80:20 split. We can also drop the loan ID since this won't be used in modelling. 

```python
#Separate target out and remove loan ID column
X = df.iloc[:, 1:].copy()
y = X.pop('Loan_Status')
```

We encode the target as a binary, applying the convention that the minority class is labelled as positive.

```python
y = y.apply(lambda x: 1 if x=='N' else 0)
```

```python
from sklearn.model_selection import train_test_split

#Split into train and test sets and stratify target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=21, stratify=y)
```

```python
#Check shapes of splits
[split.shape for split in [X_train, X_test, y_train, y_test]]
```

```python
#Check class distribution in target splits
print(y_train.value_counts(normalize=True), y_test.value_counts(normalize=True), sep='\n\n')
```

### Impute missing values

- For the categorical columns (including the loan term), we'll replace all nulls with the most frequent value. Given that 8% of the credit history values are null, this isn't ideal, but they mostly (~84%) take the value 1, so this seems reasonable in this case.
- Since the other numerical columns are right-skewed, we'll replace the nulls with the median rather than the mean.  

```python
#Check nulls
X.isna().sum()
```

```python
#Impute missing values with median
value = {'LoanAmount': X_train.loc[:, ('LoanAmount')].median()}
X_train = X_train.fillna(value=value)
X_test = X_test.fillna(value=value)  
```

```python
#Impute missing values with most frequent value
for col in categorical + ['Loan_Amount_Term']:
    value = {col: X_train.loc[:, (col)].value_counts().index[0]}
    X_train = X_train.fillna(value=value)
    X_test = X_test.fillna(value=value)
```

```python
#Check for nulls in data
print(X_train.isna().sum().sum(), X_test.isna().sum().sum())
```

### Preprocessing

First we'll recast our datatypes where appropriate.

```python
for data in X_train, X_test:
    data = data.astype({'CoapplicantIncome': 'int64', 
                        'LoanAmount': 'int64', 
                        'Loan_Amount_Term': 'int64', 
                        'Credit_History': 'int64'})
    print(data.dtypes, '\n')
```

We note the following with regard to encoding each feature. 
- Gender, Married, Education, Self_Employed and Credit_History are binary, so we'll one-hot encode these. Property_Area takes three values but there doesn't appear to be any inherent ordering so we'll also one-hot encode this and drop a redundant column.
- Dependents takes string values 0, 1, 2 and 3+. There is an ordering here so we could cast these as integers 0, 1, 2, 3, respectively, but note that this doesn't take into account the fact that a label 3 could then indicate any number of dependents from 3 upwards, while the other labels exactly match the data. Given this, we'll simply one-hot encode these as well, dropping a redundant column, as is standard.
- ApplicantIncome, CoapplicantIncome, LoanAmount and Loan_Amount_Term are all numeric and take only positive values, so we'll scale these using a MinMaxScaler().

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.preprocessing import PowerTransformer

from imblearn.pipeline import Pipeline, make_pipeline
```

```python
#numerical transforms
num_transformer = Pipeline(steps=[('power', PowerTransformer(method='yeo-johnson', standardize=False)),
                                  ('scaler', MinMaxScaler(feature_range=(0,1)))])

#categorical transforms
cat_transformer = OneHotEncoder(sparse=True, handle_unknown='ignore')

#transformer
col_transform = ColumnTransformer(transformers=[('categorical', cat_transformer, categorical),
                                                ('numerical', num_transformer, numerical[:-1]),
                                                ('scaler', MinMaxScaler(feature_range=(0,1)), [numerical[-1]])],
                                  remainder='passthrough')

```

Simple oversampling duplicates examples in the minority class and simple undersampling deletes examples in the majority class; both even out the class distribution but with undersampling we end up with a smaller training set. We'll now fit a logistic regression model using various sampling techniques and compare the results. 

### Simple sampling

First we'll use simple over- and undersampling in the pipeline. We'll use AU-ROC to measure performance; it can be overoptimistic when dealing with severely imbalanced classification problems where we have only few samples of the minority class, but should be fine for our present purposes.

```python
#Sampling and pipeline
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
```

```python
#sampling
under = RandomUnderSampler(random_state=21, sampling_strategy='majority')
over = RandomOverSampler(random_state=21, sampling_strategy='minority')
```

```python
from sklearn.linear_model import LogisticRegression
from scikitplot.metrics import roc_curve, plot_roc
from sklearn.metrics import roc_auc_score

# from sklearn.model_selection import RepeatedStratifiedKFold, GridSearchCV, cross_val_score, cross_validate
```

```python
#Define LR
lr = LogisticRegression()

#Define pipeline for non-sampled dataset
pipeline = Pipeline(steps=[('preprocessing', col_transform), ('LR', lr)])

#fit
pipeline.fit(X_train, y_train)
#get probabilities
yhat_train = pipeline.predict_proba(X_train)
yhat_test = pipeline.predict_proba(X_test)

data = [('Training', (y_train, yhat_train)), ('Test', (y_test, yhat_test))]

#Plot ROC with no sampling
fig, ax = plt.subplots(figsize=(12, 4), ncols=2)
plt.suptitle('No sampling')

for i, tup in enumerate(data):
    plot_roc(*tup[1], plot_macro=False, plot_micro=False, classes_to_plot=[1], ax=ax[i])
    ax[i].set_title(tup[0])

# Plot ROC with under-/oversampling
sampling = [('Oversampling', over), ('Undersampling', under)]

for sample in sampling:
    pipeline = Pipeline(steps=[sample, ('preprocessing', col_transform), ('LR', lr)])
    pipeline.fit(X_train, y_train)
    #get probabilities
    yhat_train = pipeline.predict_proba(X_train)
    yhat_test = pipeline.predict_proba(X_test)

    data = [('Training', (y_train, yhat_train)), ('Test', (y_test, yhat_test))]

    fig, ax = plt.subplots(figsize=(12, 4), ncols=2)

    plt.suptitle(sample[0])

    for i, tup in enumerate(data):
        plot_roc(*tup[1], plot_macro=False, plot_micro=False, classes_to_plot=[1], ax=ax[i])
        ax[i].set_title(tup[0])

```

Sampling has no effect on the model accuracy in the training set but the test set sees a slight drop in test performance for the undersampling case and a slight lift in performance for the oversampling case. This may be due to the fact that undersampling produces a smaller dataset from which to train the model on. 

```python
from collections import Counter

X_under, y_under = under.fit_resample(X_train, y_train)
X_over, y_over = over.fit_resample(X_train, y_train)

print('Target after undersampling:', Counter(y_under))
print('Target after oversampling:', Counter(y_over))
```

It's known that a combination of under- and oversampling can yield improved results too, so we'll try that next. 

### Combined under-/oversampling

To combine over- and undersampling methods, it's as simple as adding them sequentially to our pipeline. We'll first oversample the minority class and then follow this by undersampling the majority class. 

```python
#Define pipeline for combined sampling
pipeline = Pipeline(steps=[('Oversampling', over), 
                           ('Undersampling', under), 
                           ('preprocessing', col_transform), 
                           ('LR', lr)])

#fit
pipeline.fit(X_train, y_train)
#get probabilities
yhat_train = pipeline.predict_proba(X_train)
yhat_test = pipeline.predict_proba(X_test)

data = [('Training', (y_train, yhat_train)), ('Test', (y_test, yhat_test))]

#Plot ROC with no sampling
fig, ax = plt.subplots(figsize=(12, 4), ncols=2)
plt.suptitle('Combined over-/undersampling')

for i, tup in enumerate(data):
    plot_roc(*tup[1], plot_macro=False, plot_micro=False, classes_to_plot=[1], ax=ax[i])
    ax[i].set_title(tup[0])
```

In this case, we don't appear to have made any progress  the score is the same as in the oversampled case above. 

Another common oversampling technique is SMOTE (synthetic minority oversampling technique), which synthesises new data points from the minority set to be used in training. We will first run SMOTE by itself to check if we see an improvement in performance and then gridsearch through the `sampling_strategy` (which gives the ratio of the minority to majority classes) and `k_neighbors` arguments.

```python
from imblearn.over_sampling import SMOTENC
```

```python
smote = SMOTENC(sampling_strategy='minority', categorical_features=[*range(7)], random_state=21)
```

```python
#Define pipeline for SMOTE
pipeline = Pipeline(steps=[('preprocessing', col_transform), 
                           ('smote', smote),
                           ('LR', lr)])

#fit
pipeline.fit(X_train, y_train)
#get probabilities
yhat_train = pipeline.predict_proba(X_train)
yhat_test = pipeline.predict_proba(X_test)

data = [('Training', (y_train, yhat_train)), ('Test', (y_test, yhat_test))]

#Plot ROC with no sampling
fig, ax = plt.subplots(figsize=(12, 4), ncols=2)
plt.suptitle('SMOTE')

for i, tup in enumerate(data):
    plot_roc(*tup[1], plot_macro=False, plot_micro=False, classes_to_plot=[1], ax=ax[i])
    ax[i].set_title(tup[0])
```

We now see a modest improvement from 0.70 to 0.74, and also note that the training score has dropped and is within 0.06 of the test score, potentially indicating that there is less overfitting on the training set. We may see further improvement if we gridsearch through the `sampling_strategy` and `k_neighbors` arguments.

```python
ks = [*range(1,8)]
samps = [0.5, 0.6, 0.7, 0.8, 0.9]
scores = []
for samp in samps:
    for k in ks:
        smote = SMOTENC(sampling_strategy=samp, k_neighbors=k, 
                        categorical_features=[*range(7)], random_state=21)

        #Define pipeline for combined sampling
        pipeline = Pipeline(steps=[('preprocessing', col_transform), 
                                   ('smote', smote),
                                   ('LR', lr)])

        #fit
        pipeline.fit(X_train, y_train)
        #get probabilities
        yhat_train = pipeline.predict_proba(X_train)
        yhat_test = pipeline.predict_proba(X_test)

        scores.append((k, samp, roc_auc_score(y_test, yhat_test[:,1])))
        
#Print top three ROC combinations
scores.sort(key=lambda x: x[2], reverse=True)
for score in scores[:3]:
    print(f'Neighbours: {score[0]}; sampling strategy: {score[1]}; ROC AUC: {score[2]:.3f}')
```

The best scores are with 7 neighbours and a sampling strategy of 0.8 or 0.9, so either of these combinations would be fine and provide a modest lift in performance.
